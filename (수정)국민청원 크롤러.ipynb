```python
range_start = 21
range_end = 100
```

```python
import requests
from bs4 import BeautifulSoup
import time
from selenium import webdriver
import pandas as pd
import re
```

```python
url_list = []
for i in range(range_start, range_end):
    url = "https://www1.president.go.kr/petitions/" + str(i)
    url_list.append(url)
```

```python
#주어진 청원의 URL에서 정보(제목, 청원시작일, 카테고리, 서명수, 청원내용)를 추출하는 함수
def get_petition_info(url):
    while True:
        r = requests.get(url)
        time.sleep(0.5)
        if '<html><head><meta http-equiv="Refresh" content="0;' in r.text:
            continue
        else:
            break
    soup = BeautifulSoup(r.text, 'lxml')
    title = soup.title.text.split(">")[0].strip()
    if title == '국민청원':
        return '존재하지 않는 청원입니다', url.split('/')[-1]
    else:
        start_date = soup.find('ul', attrs = {'class': 'petitionsView_info_list'}).contents[3].text.replace('청원시작', '')
        category = soup.find('ul', attrs = {'class': 'petitionsView_info_list'}).contents[1].text.replace('카테고리', '')
        vote_count = soup.find('span', attrs = {'class': 'counter'}).text
        content = soup.find('div', attrs = {'class': 'View_write'}).text.strip()
        no_newline_content = re.sub(r'[\n\r]+', ' ', content)
        return start_date, category, vote_count, title, no_newline_content
```

```python
file_name = str(range_start) + '-' + str(range_end) + '_RAW.txt'
```

```python
df = pd.DataFrame(columns = {'Start_Date', 'Category', 'Vote_Count', 'Title', 'Content'})
```

```python
for url in url_list:
    with open(file_name, 'a', encoding = 'utf-8') as f:
        try:
            start_date, category, vote_count, title, no_newline_content = get_petition_info(url)
            f.write('\t\t'.join([start_date, category, vote_count, title, no_newline_content]) + '\n\n\n')
            df = df.append({'Start_Date':start_date, 'Category':category, 'Vote_Count':vote_count, 'Title':title, 'Content':no_newline_content}, ignore_index = True)
        except:
            message, petition_num = get_petition_info(url)
            f.write('\t\t'.join([message, petition_num]) + '\n\n\n')
            df = df.append({'Start_Date': petition_num, 'Category':'N/A', 'Vote_Count':'N/A', 'Title': 'N/A', 'Content':message}, ignore_index = True)
print(range_start, '번 부터 ', range_end, '번 청원까지 스크래핑 완료!')
```
