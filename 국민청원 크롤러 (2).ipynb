```python
url_list = []
for i in range(583011, 583471):
    url = "https://www1.president.go.kr/petitions/" + str(i)
    url_list.append(url)
```

```python
from bs4 import BeautifulSoup
from selenium import webdriver
import re

options = webdriver.ChromeOptions()
options.add_argument('headless')
options.add_argument('window-size=1920x1080')
options.add_argument("disable-gpu")

browser = webdriver.Chrome(executable_path=r'C:/Users/lhmlh/Desktop/chromedriver_win32/chromedriver.exe',
                          chrome_options=options)

Title = []
Contents = []
DDabong = []
Date = []
Category = []

for url in url_list:
    browser.get(url)
    html = browser.page_source
    soup = BeautifulSoup(html, 'lxml')
    title=soup.title.text.split(">")[0].strip()
    if title == "국민청원": continue
    Title.append(title)
    content = soup.find('div', style={"word-break:break-all"}).text.strip()
    Contents.append(content)
    ddabong = soup.find('span', {"class":"counter"}).text
    DDabong.append(ddabong)
    date = soup.select('ul.petitionsView_info_list>li')[1].text
    date = re.sub(r"청원시작", "", date)
    Date.append(date)
    category = soup.select('ul.petitionsView_info_list>li')[0].text
    category = re.sub(r"카테고리", "", category)
    Category.append(category)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Title":Title,"Date":Date ,"DDabong":DDabong, "Category":Category, 
                   "Contents":Contents})
df
```
