```python
df["Category"]=df.Category.apply(lambda x : x.strip())
```

```python
df=df.replace({'Category': '안전/환제'}, {'Category': '안전/환경'})
```

```python
df=df.replace({'Category': '보건복업지'}, {'Category': '보건복지'})
df=df.replace({'Category': '외교/통일/국토'}, {'Category': '외교/통일/국방'})
```

```python
set(df.Category)
```

```python
grouped = df['Vote_Count'].groupby(df['Category'])
Total_Category_Vote_Sum=grouped.sum()
df1=pd.DataFrame(Total_Category_Vote_Sum)
df1 = df1.sort_values(by="Vote_Count",ascending=False)
df1
```

# 카테고리별 청원수 Bar Plot

```python
plt.rc('font', family='Malgun Gothic')
ax=sns.barplot(data = df1, x = df1.index, y = df1.Vote_Count)
ax.figure.set_size_inches(25, 10)
plt.savefig("카테고리별_청원수.png")
```

```python
grouped = df['Nouns'].groupby(df['Category'])
Total_Category_Content_Count=grouped.count()
df2=pd.DataFrame(Total_Category_Content_Count)
df2= df2.reset_index().sort_values("Nouns",ascending=False)
df2
```

# 카테고리별 게시글 수
