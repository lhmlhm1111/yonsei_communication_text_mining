```python
from gensim.models import Word2Vec
```

### 아래 파일명만 변경해주세요

```python
with open('NOUNS_수진_20181030-20191008.txt', 'r') as f:
    docs = [x.split('\t\t')[-1].strip() for x in f.readlines()]
```

```python
docs
```

```python
docs1 = [x.split(' ') for x in docs]
```

```python
model = Word2Vec(docs1, window=3, min_count=3, size=100)
#window => 보고자 하는 양쪽 단
#min_count => 전체 corpus에서 사용 빈도수가 min_count보다 낮은 단어 제외
#size => word vector의 크기
# https://radimrehurek.com/gensim/models/word2vec.html
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#특정 단어와 문맥상 유사하게 사용된 단어
model.wv.similar_by_word('폐지',topn=20)
```

```python
#두 단어 간 문맥상 유사도
model.wv.similarity(u'폐지', u'처벌')
```
