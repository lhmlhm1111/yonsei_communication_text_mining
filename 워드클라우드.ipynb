```python
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#명사 수합해서 리스트 변수로 만들기
with open('NOUNS_수진_20181030-20191008.txt', 'r') as f: #파일명 변경해주세요
    total_docs = [x.split('\t\t')[-1].strip() for x in f.readlines()]
    
docs_nouns = [x.split(' ') for x in total_docs]

total_docs_nouns = []
for words in docs_nouns:
    total_docs_nouns.extend(words)
```

### 불용어를 추가하면서 아래 두 셀을 계속 실행하기

```python
#불용어 지정
stopwords = ['생각', '까지', '면서', '대한민국', '대해', '우리', '때문', '위해', '나라', '대로', '정부', '이나', '대한', '국민', '해주', '청원', '도록', '국가', '사람', '모든', '대통령', '한국', '문제', '정책', '이상', '정말', '내용', '아무', '보고']
```

```python
#데이터맵 생성 (불용어 및 한글자짜리 단어 제거)
data_map = Counter(x for x in total_docs_nouns if x not in stopwords and len(x) != 1)

#워드클라우드 그리기
path = '/Users/soothingni/Downloads/NanumGothicExtraBold.otf' #한글 폰트 경로 지정하기
wordcloud = WordCloud(font_path=path, max_font_size=80, background_color='white', max_words= 200)
wc = wordcloud.generate_from_frequencies(data_map)
fig = plt.figure(figsize = (10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
