```python
from ckonlpy.tag import Twitter
twitter = Twitter()
import re
import pandas as pd
import ast
```

```python
#사용자 명사 리스트 assign하기
nouns_list = ['바이오로직스', '삼바', '마리나베이', '불체자', '학생부종합', '학종', '신시장', '다가구', '다가구주택', '신은미', '더불어민주당', '각성', '문정부', '양진호', '송명빈', '다자녀', '입학금']
twitter.add_dictionary(nouns_list, 'Noun')
```

```python
df = pd.read_csv('수진_(437271-583020)_중복_결측제거.csv') #불러올 파일 제목
df = df.drop(columns = 'Unnamed: 0')
df
```

### 1. 품사 태깅 후 명사, 형용사, 동사 선택하여 외부 텍스트로 저장 (단어, 품사 모두)

```python
for k in range(len(df)):
    start_date = df.loc[k]['Start_Date']
    category = df.loc[k]['Category']
    vote_count = df.loc[k]['Vote_Count']
    text = str(df.loc[k]['Title']) + str(df.loc[k]['Content']) #제목과 본문 모두 선택
    filtered_text = re.sub(r'[^\d\w\s]', '', text.strip())
    word_pos = twitter.pos(filtered_text)
    selected_word_pos = []
    for word, pos in word_pos: #가져오고 싶은 품사 지정
        if pos == 'Noun':
            selected_word_pos.append((word, pos))
        if pos ==  'Adjective':
            selected_word_pos.append((word, pos))
        if pos ==  'Verb':
            selected_word_pos.append((word, pos))
    line = '\t\t'.join([start_date, category, str(vote_count), str(selected_word_pos)])
    with open('43271-583020_POS_tagged.txt', 'a') as f: #저장할 파일 제목
        f.write(line + '\n\n\n')
```

### 2. 명, 형, 동 태깅된 외부 텍스트 불러와서 특정 품사 단어만 선택한 후 외부 텍스트로 저장 (단어만, 품사x)

```python
df = pd.read_csv('43271-583020_POS_tagged.txt', sep = '\t\t', names = ['Start_Date', 'Category', 'Vote_Count', 'Word_POS']) #위에서 저장한 파일 제목
```

```python
def extract_word_by_pos(cell, POS):
    cell = ast.literal_eval(cell)
    selected_words = [word for word, pos in cell if pos == POS]
    return selected_words
```

```python
#명사만 추출하기
for k in range(len(df)):
    row = df.loc[k]
    start_date = row['Start_Date']
    category = row['Category']
    vote_count = row['Vote_Count']
    selected_words = extract_word_by_pos(row['Word_POS'], 'Noun') #추출하고자 하는 품사 지정해주기
    word_string = ""
    for word in selected_words:word_string += word + ' '
    line = '\t\t'.join([start_date, category, str(vote_count), word_string])
    with open('파일제목.txt', 'a') as f:
        f.write(line + '\n\n\n')
```
