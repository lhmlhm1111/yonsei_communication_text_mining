```python
import requests
from bs4 import BeautifulSoup
import time
from selenium import webdriver
import pandas as pd
```

```python
#청원 URL을 모으는 함수; 인자는 첫 페이지의 URL
def get_urls(url):
    #for문이 돌아갈 range 구하기
    driver.get(url)
    count = driver.find_element_by_xpath('//*[@id="cont_view"]/div[2]/div/div/div[2]/div[2]/div[4]/div/div[2]/div[2]/ul/li[1]/div/div[1]').text
    range_num = int(count)//15 + 3
    url_list = []
    for k in range(1, range_num):
        url = 'https://www1.president.go.kr/petitions/?c=0&only=1&page=' + str(k) + '&order=1'
        driver.get(url)
        time.sleep(1)
        petition_board = driver.find_element_by_xpath('//*[@id="cont_view"]/div[2]/div/div/div[2]/div[2]/div[4]/div/div[2]/div[2]/ul')
        petitions_list = petition_board.find_elements_by_class_name('cb.relpy_w')
        for i in petitions_list:
            url_list.append(i.get_attribute('href'))
    return url_list

#주어진 청원의 URL에서 정보(제목, 청원시작일, 카테고리, 서명수, 청원내용)를 추출하는 함수
def get_petition_info(url):
    while True:
        r = requests.get(url)
        time.sleep(0.5)
        if '<html><head><meta http-equiv="Refresh" content="0;' in r.text:
            continue
        else:
            break
    soup = BeautifulSoup(r.text, 'lxml')
    title = soup.find('h3', attrs = {'class': "petitionsView_title"}).text
    start_date = soup.find('ul', attrs = {'class': 'petitionsView_info_list'}).contents[3].text.replace('청원시작', '')
    category = soup.find('ul', attrs = {'class': 'petitionsView_info_list'}).contents[1].text.replace('카테고리', '')
    count = soup.find('span', attrs = {'class': 'counter'}).text
    content = soup.find('div', attrs = {'class': 'View_write'}).text.strip()
    return title, start_date, category, count, content
```

```python
driver = webdriver.Chrome('/Users/soothingni/Documents/python_data/chromedriver')

url_list = get_urls('https://www1.president.go.kr/petitions/?c=0&only=1&page=1&order=1')
```

```python
current_petition_result = pd.DataFrame(columns = {'Title', 'Start_Date', 'Category', 'Count', 'Content'})

for k in range(len(url_list)):
    title, start_date, category, count, content = get_petition_info(url_list[k])
    current_petition_result = current_petition_result.append({'Title':title, 'Start_Date':start_date, 'Category':category, 'Count':count, 'Content':content}, ignore_index = True)
```

```python
#불필요한 특수문자 제거
import re
for k in range(len(current_petition_result)):
    current_petition_result['Title'][k] = re.sub(r'[^0-9a-zA-Zㄱ-ㅣ가-힣#!?.\s\*]', '', current_petition_result['Title'][k]).replace('\xa0', ' ').replace('\u2003', ' ')
    current_petition_result['Content'][k] = re.sub(r'[^0-9a-zA-Zㄱ-ㅣ가-힣#!?.\s\*]', '', current_petition_result['Content'][k]).replace('\xa0', ' ').replace('\u2003', ' ')

#csv로 저장
current_petition_result.to_csv('진행중인_청원_191105.csv', encoding = 'euc-kr')    
```
