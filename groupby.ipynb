```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
%matplotlib inline
```

```python
df=pd.read_csv("학민_(437271_583020)_중복_결측제거.csv")
df=df.drop("Unnamed: 0", axis=1)
df
```

```python
df["Vote_Count"]=df.Vote_Count.apply(lambda x : float(x.strip().replace(" ","")))
```

```python
df["Vote_Count"]=df.Vote_Count.apply(lambda x : int(x))
```

```python
df["Start_Date"]=df.Start_Date.apply(lambda x : str(x))
df["Start_Date"]=df["Start_Date"].astype('datetime64[ns]')
```

```python
df.dtypes
```

```python
grouped = df['Vote_Count'].groupby(df['Category'])
```

```python
Total_Category_Vote_Count=grouped.count()
Total_Category_Vote_Count
```

```python
Total_Category_Vote_Sum=grouped.sum()
Total_Category_Vote_Sum
```

```python
Total_Category_Vote_Mean=grouped.mean()
Total_Category_Vote_Mean
```
